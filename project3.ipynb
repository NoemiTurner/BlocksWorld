## Goal State
Our goal in Blocks World is to have a stack of 9 blocks that are in sequential order (block 9 on top and 1 on the bottom). 
<br>
<br>
Example: The blocks would look like this.
> __ Position n __ 
> 1
> 2
> 3
> 4
> 5
> 6
> 7
> 8
> 9

## Create function to display the current positions of the blocks graphically

```python
# pretty print helper function
def pretty_print(position_list):
    for i, position in enumerate(position_list):
        print("__ Position" ,i+1, "__  [", end="")
        for i, block in enumerate(position):
            if i == len(position) - 1:
                print(block, end="")
            else:
                print(str(block) + ",", end="")
        print("]")
        print()
```

## Generate 9 blocks numbered 1-9 and then initially distribute the blocks randomly over 9 positions

```python
import random
blocks = [1,2,3,4,5,6,7,8,9]

# create a 2D list that represents the positions the blocks can be placed inside
start_positions = [[] for i in range(len(blocks))]

# randomly place the 3 blocks in the different positions
for i in range(len(start_positions)):
    pos = random.randint(0,8) # choose a random position to place the block
    start_positions[pos].append(blocks[i]) # put the block in the position

goal_state = [1,2,3,4,5,6,7,8,9]

positions = start_positions 

print("Starting Poisitions of the Blocks:\n")
pretty_print(positions)
```

## Move the blocks around randomly until they are all in one column and in order

```python
import time
startTime = time.time()

num_moves = 0

flag = True

while flag == True:
    if goal_state in positions:
        print("Solved!!")
        print("Ending Positions:")
        pretty_print(positions)
        flag = False
        
    pos = random.randint(0,8)
    while positions[pos] == []:
        pos = random.randint(0,8)

    if positions[pos] != []:
        block = positions[pos].pop(-1)

        new_location = random.randint(0,8)
        while positions[new_location] == positions[pos]:
            new_location = random.randint(0,8)
 
        positions[new_location].append(block)
        num_moves += 1
        # print("___________________________________")
        # print()
        # pretty_print(positions)

print(num_moves)
executionTime = (time.time() - startTime)
print('Execution time in seconds: ' + str(executionTime))
```
